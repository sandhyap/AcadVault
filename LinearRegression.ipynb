Define the cost function
Write a Python function to compute the cost function and test it out on a few different pairs of
values for and .

```python

def cost_function(X,y,theta):
    #m = len(y)
    m=20
    J=0 # initialize loss to zero
    
    # reshape theta
    theta=theta.reshape(2,1)
    
    # calculate the hypothesis - y_hat
    h_x = np.dot(X,theta)
    
    # subtract y from y_hat, square and sum
    error_term = sum((h_x - y)**2)
    
    # divide by twice the number of samples - standard practice.
    loss = error_term/(2*m)
    
    return loss
```

```python
 # cost function is implemented in the file gd
theta=np.array([[0],[1]])
theta1=np.array([[2],[-1]])
theta2=np.array([[-1],[2]])
print (cost_function(x, y,theta)) #this is the MSE cost function
print (cost_function(x, y,theta1))
print (cost_function(x, y, theta2))
```

```python
theta0_vals = np.linspace(-20, 20)
theta1_vals = np.linspace(-20, 20)
print(theta0_vals.shape)
THETA0, THETA1 = np.meshgrid(theta0_vals, theta1_vals)
print(THETA0.shape)
```

```python
def mycost(theta0, theta1):
  theta=np.array([[theta0],[theta1]])
  return cost_function(x, y, theta)
```

```python
mycost_vectorized = np.vectorize(mycost)
J_SURF = mycost_vectorized(THETA0,THETA1)
J_SURF.shape
```

```python
def gradient(x, y, theta_0, theta_1):
  theta=np.array([[theta_0],[theta_1]])
  theta=theta.reshape(2,1)
  return (-2 * (x.transpose()).dot (y - x.dot(theta) ))/20
```

Use gradient descent to find the best
parameters

```python
#gradient descent algorithm function
def gradient_descent_algo(theta_0, theta_1, max_iters, alpha):
  theta_0_history = [theta_0]
  theta_1_history = [theta_1]
  plt.figure(figsize=(10,4))
  for i in range(0, max_iters):
    print('iter=',i)
    # Here is the gradient step
    d0, d1 = gradient(x, y, theta_0, theta_1)
    print(d0," ",d1)
    theta_0 =float( theta_0 - alpha*d0)
    theta_1 =float( theta_1 - alpha*d1)
    # The rest is plotting
    clear_output(wait=True)
    plt.figure(figsize=(10,4))
    theta_0_history.append(theta_0)
    theta_1_history.append(theta_1)
    plt.subplot(1, 2, 1)
    plt.contour(THETA0, THETA1, J_SURF, 20)
    plt.xlabel('Intercept theta_0')
    plt.ylabel('Slope theta_1')
    plt.plot(theta_0_history, theta_1_history)
    plt.plot( theta_0, theta_1, marker='o')
    plt.xlim([-10, 10])
    plt.ylim([-10, 10])
    plt.subplot(1, 2, 2)
    plt.plot(x1, y1, marker='o', linestyle='none')
    plt.plot( x_test, theta_0 + theta_1 * x_test)
    plt.xlim([-6, 6])
    plt.ylim([-10, 10])
    plt.show()
    time.sleep(0.2)
    print("Final parameters:\ntheta0=", theta_0, "\ntheta1=", theta_1)
```

```python
# Run gradient descent (Settings 1)
theta_0 = 1
theta_1 = 2
max_iters = 20
alpha = 0.1
gradient_descent_algo(theta_0, theta_1, max_iters, alpha)
```
